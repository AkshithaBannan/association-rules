```python
#applying apriori algorithm with a lower support threshold and generating association rules
import warnings
warnings.filterwarnings("ignore")

frequent_itemsets = apriori(one_hot_encoded, min_support=0.02, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

print("Frequent Itemsets:")
print(frequent_itemsets)
```

```python
#applying association rules with lower confidence threshold
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules
```

```python
print("\nAssociation Rules:")
print(rules)
```

```python
print("Number of generated rules:", len(rules))
```

```python
#converting lift column to numerical datatype
rules['lift'] = pd.to_numeric(rules['lift'])

# Display the top 10 rules with the highest lift
top_rules_lift = rules.nlargest(10, 'lift')
print("\nTop 10 rules with the highest lift:")
print(top_rules_lift)
```

```python
#converting confidence to numerical data
rules['confidence'] = pd.to_numeric(rules['confidence'])

# Display the top 10 rules with the highest confidence
top_rules_confidence = rules.nlargest(10, 'confidence')
print("\nTop 10 rules with the highest confidence:")
print(top_rules_confidence)
```

```python
#rules with lift > 1 and confidence > 0.7
high_confidence_rules = rules[(rules['lift'] > 1) & (rules['confidence'] > 0.7)]
print("\nRules with lift > 1 and confidence > 0.7:")
print(high_confidence_rules)
```

```python
#Lift and importance of lift in association rules

#It is a metric used to mine the relationship between the two items.
#A high lift value indicates that the association rule is more significant as two items are highly dependent on each other.
```

```python
#Support : It is a measure of number of times the item set appears in the dataset.
#          Calculated by the number of transactions containing an item set by the total number of transactions.
```

```python
#Confidence : It is a measure of likelihood that an itemset will appear if another itemset appears.
#             Calculated by dividing the number of transactions containing both itemsets by number of transactions containing
#             first itemset.
```

```python
#Limitations of using association rules :-
#Poor performance
#produces boring rules
#already have many discovered rules
#noise and performance effects
#not suitable for heterogeneous data
#sensitive to thresholds
#generating too many rules for large datasets which makes it messy and hard to find important patterns
```
